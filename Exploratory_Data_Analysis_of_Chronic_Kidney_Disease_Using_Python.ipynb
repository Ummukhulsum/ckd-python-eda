Exploratory Data Analysis of Chronic Kidney Disease Using Python

 Introduction 
Chronic Kidney Disease (CKD) is a major public health challenge worldwide, particularly in low- and middle-income regions where late diagnosis and limited healthcare resources contribute to high mortality rates. CKD often progresses without obvious symptoms until significant kidney damage has occurred. Clinical factors such as age, blood pressure, diabetes, hypertension, and abnormal laboratory values play a critical role in the development and progression of the disease.
The aim of this project is to perform an exploratory data analysis (EDA) on a chronic kidney disease dataset using Python in order to identify patterns, trends, and relationships among demographic and clinical variables. This analysis demonstrates how data science techniques can be applied to healthcare data to support better understanding of chronic kidney disease.

Data Cleaning and Processing

Medical datasets often contain missing values. To ensure accurate analysis:
Numerical variables were filled using the median.
Categorical variables were filled using the mode.

```python
import numpy as np

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Exploratory Data Analysis (EDA)
Age Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['age'], bins=20)
plt.title("Age Distribution of CKD Patients")
plt.show()
```

Serum Creatinine vs Age

```python
sns.scatterplot(x='age', y='sc', data=df)
plt.title("Age vs Serum Creatinine")
plt.show()
```

Hypertension Prevalence

```python
df['htn'].value_counts().plot(kind='bar')
plt.title("Hypertension Among CKD Patients")
plt.show()
```

Key Findings
CKD is more prevalent among older individuals.
Many patients have elevated serum creatinine and blood urea levels.
Hypertension and diabetes are common among CKD patients.
Low hemoglobin levels indicate a high occurrence of anemia.
